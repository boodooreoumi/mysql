```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# 그래프에 한글설정
plt.rcParams['font.family']='Malgun Gothic'

# 유니코드에서 그래프에 마이너스 기호 깨지는 문제 해결
plt.rcParams['axes.unicode_minus']=False
```

```python
blood_type=['A','B','O','AB']
Personnel=[111901,87066,86804,36495]
```

```python
# 기본 파이그래프 그리기
plt.pie(Personnel)
plt.show()
```

```python
blood_type=['A','B','O','AB']
Personnel=[111901,87066,86804,36495]

plt.pie(Personnel, labels=blood_type, labeldistance=1.2, autopct='%.1f%%',
       pctdistance=0.7, explode=[0.1,0,0,0],
       colors=['lightcoral','gold','greenyellow','skyblue'],
       startangle=90, counterclock=False,
       radius=1.3)
plt.show()

# 1. labeldistance=1.2 (라벨의 거리)
# labeldistance는 각 레이블이 원형 차트의 중심에서 얼마나 떨어져 있을지를 설정하는 값입니다. 
# 1.2는 레이블이 차트의 중심에서 약간 바깥쪽에 위치하도록 합니다.

# 2. autopct='%.1f%%' (백분율 표시)
# autopct는 각 조각에 대한 백분율을 표시하는 형식입니다. 
# %.1f%%는 소수점 첫째 자리까지 백분율을 표시하라는 뜻입니다. 예를 들어, 인원 수 비율이 0.25이면 25.0%로 표시됩니다.

# 3. pctdistance=0.7 (백분율의 위치)
# pctdistance는 백분율이 차트의 원형 안에서 어디에 위치할지 결정하는 값입니다. 0.7은 차트의 중심에서 70% 거리 위치에 백분율이 표시되도록 합니다.

# 4. explode=[0.1, 0, 0, 0] (조각의 분리)
# explode는 각 조각을 얼마나 밖으로 "튀어나오게" 할지 설정하는 리스트입니다. 
# 0.1은 'A' 혈액형의 조각을 약간 튀어나오게 하고, 나머지 조각들은 튀어나오지 않게 설정합니다. 
# 이 효과는 시각적으로 특정 항목을 강조하는 데 유용합니다.

# 5. colors=['lightcoral', 'gold', 'greenyellow', 'skyblue'] (조각의 색상)
# colors는 각 조각의 색상을 지정합니다. 순서대로 각 혈액형에 대해 색상을 지정해주며, 
# 이 예에서는 'A' 혈액형은 lightcoral, 'B' 혈액형은 gold, 'O' 혈액형은 greenyellow, 'AB' 혈액형은 skyblue 색으로 설정됩니다.

# 6. startangle=90 (시작 각도)
# startangle은 원형 차트가 그려지는 시작 각도를 설정합니다. 
# 기본값은 0인데, 90으로 설정하면 원형 차트가 시계방향으로 90도 회전한 상태로 그려지며, 첫 번째 섹션이 12시 방향에서 시작하게 됩니다.

# 7. counterclock=False (시계 반대 방향으로 회전)
# counterclock가 False이면 차트가 시계방향으로 그려집니다. 기본값은 True로 시계반대방향입니다.

# 8. radius=1.3 (차트의 크기)
# radius는 원형 차트의 크기를 조절하는 값입니다. 기본값은 1이고, 1.3으로 설정하여 차트의 크기를 약간 확대합니다.
```

```python
plt.figure(figsize=(5,5), facecolor='ivory', edgecolor='gray', linewidth=2)

plt.pie(Personnel, labels=blood_type, labeldistance=1.2, autopct='%.1f%%',
       pctdistance=0.7, explode=[0.1,0,0,0],
       colors=['lightcoral','gold','greenyellow','skyblue'],
       startangle=90, counterclock=False,
       radius=1, wedgeprops={'ec':'k','lw':2,'ls':':','width':0.7},
       textprops={'fontsize':12,'color':'b','rotation':0})

plt.title('2019년 병역판정검사 : 혈액형 분포', size=15)
plt.show()
```
