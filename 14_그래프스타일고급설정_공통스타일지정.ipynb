```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# rcParams : 그래프를 구성하는 공통속성을 지정

plt.rcParams['font.family']='Malgun Gothic'
plt.rcParams['axes.unicode_minus']=False
```

```python
# 1. figure 크기
plt.rcParams['figure.figsize']=(9,4)
```

```python
# 2. axes - 그래프설정
# 그래프 테두리 두께
plt.rcParams['axes.linewidth']=2
# 그래프 테두리 색
plt.rcParams['axes.edgecolor']='navy'
# 그래프 바탕색
plt.rcParams['axes.facecolor']='ghostwhite'
# 그리드 표시
plt.rcParams['axes.grid']=True
```

```python
# 3. 그래프 제목
plt.rcParams['axes.titlecolor']='navy'
plt.rcParams['axes.titlesize']=15
plt.rcParams['axes.titleweight']='bold'
```

```python
# 4. 레이블
plt.rcParams['axes.labelcolor']='gray'
plt.rcParams['axes.labelsize']=12
```

```python
# 5. 그리드
plt.rcParams['grid.alpha']=0.3
plt.rcParams['grid.color']='skyblue'
plt.rcParams['grid.linestyle']='--'
```

```python
# 6. 눈금
# x축 눈금선이 상단에 나타날지 여부를 설정하는 옵션이며,
# 기본적으로 x축의 눈금선이 하단에만 표시되지만
# 이설정을 통해 상단에도 눈금선이 추가된다
plt.rcParams['xtick.top']=True

# y축 눈금선이 우측에 나타날지 여부를 설정하는 옵션이며,
# 기본적으로 y축의 눈금서니 좌측에만 표시되지만
# 이설정을 통해 우측에도 눈금선이 추가된다
plt.rcParams['ytick.right']=True

# x축 눈금선의 방향을 설정하는 옵션이며, 그래프 안쪽으로 향하게 된다
plt.rcParams['xtick.direction']='in'

# y축 눈금선의 방향을 설정하는 옵션이며, 그래프 안쪽으로 향하게 된다
plt.rcParams['ytick.direction']='in'

# 주요x축 눈금선의 길이를 설정하는 옵션으로 10포인트로 설정된다
plt.rcParams['xtick.major.size']=10

# 주요y축 눈금선의 길이를 설정하는 옵션으로 10포인트로 설정된다
plt.rcParams['ytick.major.size']=10
```

```python
# 7. 선
plt.rcParams['lines.linewidth']=3
plt.rcParams['lines.linestyle']='--'
plt.rcParams['lines.marker']='o'
```

```python
# 8. 패턴
plt.rcParams['hatch.linewidth']=3
plt.rcParams['hatch.color']='w'
```

```python
# 9. 박스플롯
plt.rcParams['boxplot.boxprops.color']='b'
plt.rcParams['boxplot.boxprops.linewidth']=2

# 박스플롯 최대값, 최소값
plt.rcParams['boxplot.capprops.color']='r'
plt.rcParams['boxplot.capprops.linewidth']=2

# 박스플롯 평균
plt.rcParams['boxplot.showmeans']=True
```

```python
# 선 그래프
df1 = pd.DataFrame({'요일':['월','화','수','목','금','토','일'],
                   '매출액':[10000,9000,11000,8000,13000,15000,14000]})

plt.plot(df1['요일'], df1['매출액'])
plt.title('요일별 매출액')
plt.xlabel('요일')
plt.ylabel('매출액(만원)')
plt.show()
```

```python
# 막대그래프
plt.bar(df1['요일'],df1['매출액'],hatch='/')
plt.title('요일별 매출액')
plt.xlabel('요일')
plt.ylabel('매출액(만원)')
plt.show()
```
