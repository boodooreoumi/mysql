```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 그래프에 한글폰트 설정
plt.rcParams['font.family']='Malgun Gothic'
# 그래프에 마이너스 기호 깨지는 문제 해결
plt.rcParams['axes.unicode_minus']=False
```

```python
# 그리드 스타일 : 라인스타일(--), 투명도(0.3)
plt.rcParams['grid.linestyle']='--'
plt.rcParams['grid.alpha']=0.3
```

```python
# 수평선 그리기
plt.plot([1,2,3,4], 'ko')

plt.hlines(2, 0, 1, color='b', alpha=0.5, ls=':') # y=2, xmin=0, xmax=1
plt.hlines(3,0,2,color='darkorange',ls='-.') # y=3, xmin=0, xmax=3

plt.show()
```

```python
# 수직선그리기
plt.plot([1,2,3,4], 'ko')
plt.vlines(1,0,2,color='b',alpha=0.5,ls=':')
plt.vlines(2,0,3,color='darkorange',alpha=0.5,ls=':')

plt.show()
```

```python
# 예제 : 요일별 판매 테이블수

import seaborn as sns
tips = sns.load_dataset('tips')
tips.head(3)
```

```python
# 요이별 테이블수
s = tips['day'].value_counts()
s
```

```python
plt.figure(figsize=(15,3))
plt.barh(s.index, s.values)
plt.xticks(range(0,91,1), rotation=90)
plt.grid(axis='x')

plt.axvline(s['Fri'], color='k')
plt.axvline(s['Thur'], color='k')
plt.axvline(s['Sun'], color='k')
plt.axvline(s['Sat'], color='k')

plt.show()

# axvline는 전체플롯에 수직선을 그리는 함수이고,
# vlines는 여러 개의 수직선을 동시에 그릴 수 있는 함수이다.
# 각 수직선의 시작점과 끝점을 개별적으로 설정할 수 있다
```
