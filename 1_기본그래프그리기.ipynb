```python
import matplotlib.pyplot as plt
```

```python
# 리스트 데이터로 그리기
data = [1,3,5,9,7]
plt.plot(data)
```

```python
# 시리즈 데이터로 그리기
import pandas as pd
s = pd.Series([1,3,5,9,7])
plt.plot(s)
```

```python
# 인덱스변경하기
ss=pd.Series([1,3,5,9,7], index=[10,20,30,40,50])
plt.plot(ss)
```

```python
# 문자인덱스로 변경하기
sss = pd.Series([1,3,5,7,9], index=['a','b','c','d','e'])
plt.plot(sss)
```

```python
# x축, y축 데이터로 그리기
x = [1,5,6,9,10]
y = [1,5,3,9,7]
plt.plot(x,y)
```

```python
# 데이터프레임 데이터로 그리기
sss = pd.DataFrame({'x':[1,5,6,9,10],'y':[1,5,3,9,7]})
sss
```

```python
# 그래프 그리기
x = sss['x']
y = sss['y']
plt.plot(x,y)
```

```python
# 시리즈 데이터로 그리기
s=pd.Series([1,5,3,9,7])
s
plt.plot(s)
plt.show()
```

```python
plt.plot(s.index, s.values)
plt.show()
```
