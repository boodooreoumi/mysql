```python
import matplotlib.pyplot as plt
```

```python
# 한글폰트 사용하기위해 폰트 확인하기
import matplotlib.font_manager as fm
[f.name for f in fm.fontManager.ttflist if 'Malgun Gothic' in f.name]
```

```python
# 한글폰트 지정 및 음수표현
plt.rcParams['font.family']='Malgun Gothic'
plt.rcParams['axes.unicode_minus']=False
```

```python
x=['봄','여름','가을','겨울']
y=[19,38,15,-10]
plt.plot(x,y)
plt.show()
```

```python
# 색상
x = ['봄','여름','가을','겨울']
y = [19,38,15,-10]
plt.plot(x,y,color='r')
plt.show()
```

```python
# 마커
plt.plot(x,y,color='r',marker='*',ms='10',mec='b',mew='3',mfc='y', ls=':', lw='10')
# marker=마커종류, ms=마커사이즈, mec=마커선색깔, mew=마커선굵기, mfc=마커내부색깔, ls=linestype, lw=선굵기
```

```python
# 선색상이 블랙
plt.plot(x,y,'k')
```

```python
# 마커종류가 다이아몬드
plt.plot(x,y,'D')
```

```python
# 선스타일
plt.plot(x,y,':')
```

```python
# 마커색상은 빨강, 마커종류는 다이아몬드
plt.plot(x,y,'rD')
```

```python
# 색상은 빨강, 선스타일은 점선
plt.plot(x,y,'r:')
```

```python
# 마커는 다이아몬드, 선스타일은 점선
plt.plot(x,y,'D:')
```

```python
plt.plot(x,y,'rD:',ms='10',mec='r',mfc='b',lw='2')
```
