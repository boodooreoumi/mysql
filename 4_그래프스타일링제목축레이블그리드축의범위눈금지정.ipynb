```python
import matplotlib.pyplot as plt
```

```python
# 한글폰트사용
# 폰트 확인하기
import matplotlib.font_manager as fm
[f.name for f in fm.fontManager.ttflist if 'Malgun Gothic' in f.name]
```

```python
# 한글폰트 지정
plt.rcParams['font.family']= 'Malgun Gothic'
plt.rcParams['axes.unicode_minus']=False
```

```python
# 샘플데이터
# 월별 몸무게 데이터프레임

import pandas as pd
df = pd.DataFrame({'월':[1,2,3,4,5],'몸무게':[80,70,75,73,70]})
df
```

```python
# 그래프 그리기
x = df['월']
y = df['몸무게']
plt.plot(x,y,'ro-',mec='b',mfc='b',lw=3,ms=7)
plt.title('월별몸무게변화', loc='right', pad=10, color='r', fontsize=20, fontweight='bold', fontfamily='Malgun Gothic')
plt.xlabel('월',loc='right',color='b',fontsize=12) # loc = left, right
plt.ylabel('몸무게(kg)', loc='top', labelpad=10, fontsize=12) # bottom, center, top
plt.grid(axis='x',color='skyblue',alpha=0.5,ls=':')
plt.show()
```

```python
# 축의 범위와 눈금 지정
df = pd.DataFrame({'월':[1,2,3,4,5],'몸무게':[80,78,75,73,70]})
df
```

```python
# 그래프 그리기
x=df['월']
y=df['몸무게']
plt.plot(x,y,'bo-')
plt.title('월별 몸무게 변화',size=15,color='b',fontweight='bold')
plt.xlabel('월')
plt.ylabel('몸무게(kg)')
plt.grid(ls=':')
plt.show()
```

```python
# 축의 범위 추가하기

xticks_lable = ['','1월','2월','3월','4월','5월','6월','7월','8월','9월','10월','11월','12월','']
yticks_lable = ['0kg','10kg','20kg','30kg','40kg','50kg','60kg','70kg','80kg','90kg','100kg']

x=df['월']
y=df['몸무게']
plt.plot(x,y,'bo-')
plt.title('월별 몸무게 변화',size=15,color='b',fontweight='bold')
plt.xlabel('월')
plt.ylabel('몸무게(kg)')

# 축의 범위
plt.xlim(1,12)
plt.ylim(0,100)

# 틱설정
plt.xticks(range(0,14,1), labels=xticks_lable)
plt.yticks(range(0,110,10), labels=yticks_lable)

# 틱스타일 지정
plt.tick_params(width=2, color='g')

plt.grid(ls=':')
plt.show()
```

```python
# 여러 개의 그래프와 범례
df1 = pd.DataFrame({'월':[1,2,3,4,5],'몸무게':[80,78,75,73,70]})
df2 = pd.DataFrame({'월':[1,2,3,4,5],'몸무게':[60,62,59,55,54]})
```

```python
df1
```

```python
df2
```

```python
# 여러개의 그래프 각각 그리기
plt.plot(df1['월'],df1['몸무게'])
plt.show()

plt.plot(df2['월'],df2['몸무게'])
plt.show()
```

```python
def fn_xtick(x):
    x = str(x)
    if x=='0' or x=='13':
        return ''
    else:
        return x+'월'
    
xtick_label = pd.Series(range(0,14,1))
xtick_label = xtick_label.apply(fn_xtick)
xtick_label
```

```python
def fn_ytick(x):
    x = str(x)
    return x+'kg'

ytick_label = pd.Series(range(30,95,5))
ytick_label = ytick_label.apply(fn_ytick)
ytick_label
```

```python
# 두 그래프 한번에 그리기 + 범례
plt.plot(df1['월'],df1['몸무게'], label='James')
plt.plot(df2['월'],df2['몸무게'], label='Amy')
plt.legend(loc='upper right', ncol=2)

# 그래프 제목
plt.title('월별 몸무게 변화', size=15)

# 축라벨
plt.xlabel('월', fontsize=12)
plt.ylabel('몸무게', fontsize=12)

# 틱
plt.xticks(range(0,14,1), labels=xtick_label)
plt.yticks(range(30,95,5), labels=ytick_label)

# 그리드
plt.grid(ls=':')

plt.show()
```
