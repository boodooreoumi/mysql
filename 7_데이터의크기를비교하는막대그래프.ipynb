```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 그래프에 한글 설정
plt.rcParams['font.family']= 'Malgun Gothic'
# 그래프에 마이너스 기호 깨지는 문제 해결
plt.rcParams['axes.unicode_minus']=False
```

```python
# 샘플데이터
# 어느 두 매장의 요일별 평균매출액
df1 = pd.DataFrame({'요일':['월','화','수','목','금','토','일'],
                   '매출액':[10000,9000,11000,8000,13000,15000,14000]})
df1
```

```python
df2 = pd.DataFrame({'요일':['월','화','수','목','금','토','일'],
                   '매출액':[9000,9500,13000,7000,12000,14000,11000]})
```

```python
# 막대그래프
# 세로막대그래프
plt.bar(df1['요일'],df1['매출액'])
plt.show()
```

```python
# 가로막대그래프
plt.barh(df1['요일'],df1['매출액'])
plt.show()
```

```python
# 세로막대그래프_막대폭지정
plt.bar(df1['요일'],df1['매출액'],width=0.6)
plt.show()
```

```python
# 가로막대그래프_막대폭지정
plt.barh(df1['요일'],df1['매출액'],height=0.6)
plt.show()
```

```python
# 막대그래프 막대마다 색상지정 + 막대테두리 + 막대패턴지정 + 패턴의 밀도저장
plt.bar(df1['요일'],df1['매출액'],width=0.4, color=['r','g','y','b','navy','pink','violet'],
       edgecolor='gray', linewidth=1, hatch='****')
plt.show()
```

```python
# 막대마다 다른 패턴지정
# set_hatch()
bars=plt.bar(df1['요일'],df1['매출액'],width=0.4,color=['r','g','y','b','navy','pink','violet'],
       edgecolor='gray', linewidth=1)
bars[0].set_hatch('.')
bars[1].set_hatch('+')
bars[2].set_hatch('/')
bars[3].set_hatch('x')
bars[4].set_hatch('+')
bars[5].set_hatch('++')
bars[6].set_hatch('--')
```

```python
# 막대위치 지정
plt.bar(df1['요일'],df1['매출액'],width=-0.4,align='edge')
plt.show()
```

```python
# 막대위치 지정
plt.bar(df1['요일'],df1['매출액'],width=0.4,align='edge')
plt.show()
```

```python
# 두개의 막대그래프 그리기
plt.bar(df1['요일'],df1['매출액'],width=-0.4,align='edge') # 중심왼쪽
plt.bar(df2['요일'],df2['매출액'],width=0.4,align='edge')  # 중심오른쪽
plt.title('두 매장의 요일별 매출액 평균',size=15)
plt.grid(axis='y', ls=':')
```

```python
plt.barh(df1['요일'],df1['매출액'],height=-0.4,align='edge')
plt.barh(df2['요일'],df2['매출액'],height=0.4,align='edge')
plt.title('두 매장의 요일별 매출액 평균', size=15)
plt.grid(axis='x',ls=':')
```
