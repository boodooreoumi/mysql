```python
import matplotlib.pyplot as plt

# 그래프에 한글 설정
plt.rcParams['font.family']= 'Malgun Gothic'
# 그래프에 마이너스 기호 깨지는 문제 해결
plt.rcParams['axes.unicode_minus']=False
```

```python
# 데이터
x = [1,5,6,9,10]
y = [1,5,3,9,7]
plt.scatter(x,y)
```

```python
plt.plot(x,y,'D')
```

```python
# 점의 크기 색상 설정
plt.scatter(x,y,s=[100,200,50,60,30], c=['r','b','y','g','k'])
```

```python
# 컬러맵으로 색깔지정
plt.scatter(x,y,c=y,cmap='Oranges')
plt.colorbar()
```

```python
# colormap의 종류
plt.colormaps()
```

```python
# 산점도 예제
# 데이터 준비
import seaborn as sns
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.shape
```

```python
# 지불금액과 팁의 관계
# 지불금액이 클수록 팁도 많이 줄까?
plt.scatter(tips['total_bill'],tips['tip'])
```

```python
# 점의 크기로 테이블 인원 표시
plt.scatter(tips['total_bill'],tips['tip'],s=tips['size']*15,alpha=0.5)
```

```python
# 점의 색으로 성별 표시
# 성별에 따른 color 컬럼 추가

def set_color(x):
    if x=='Male':
        return 'blue'
    elif x=='Female':
        return 'red'

tips['color']=tips['sex'].apply(set_color)
tips.head(5)
```

```python
plt.scatter(tips['total_bill'],tips['tip'],s=tips['size']*15,alpha=0.5,c=tips['color'])
```
